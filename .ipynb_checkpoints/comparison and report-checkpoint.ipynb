## compare cell and t-test 

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

# Load data
df1 = pd.read_csv("my_goals_results.csv")
df2 = pd.read_csv("my_goals_results_second.csv")

# Basic summary
def summarize(df):
    total = len(df)
    successes = (df["status"] == "success").sum()
    failures = total - successes
    avg_duration = df["duration_sec"].mean()
    return total, successes, failures, avg_duration

total1, success1, fail1, avg1 = summarize(df1)
total2, success2, fail2, avg2 = summarize(df2)

print("=== Summary Statistics ===")
print(f"First Implementation:  {total1} goals | Success: {success1} | Failed: {fail1}")
print(f"Second Implementation: {total2} goals | Success: {success2} | Failed: {fail2}\n")

print("Average Duration (sec):")
print(f"  First:  {avg1:.2f}")
print(f"  Second: {avg2:.2f}")

# T-test
t_stat, p_val = ttest_ind(df1["duration_sec"], df2["duration_sec"])
print("\n=== t-test Results ===")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value:     {p_val:.4f}")
if p_val < 0.05:
    print("🔍 Significant difference in durations (p < 0.05)")
else:
    print("ℹ️ No significant difference in durations (p >= 0.05)")

# Plot durations
plt.figure(figsize=(10, 5))
plt.plot(df1["duration_sec"].values, label="First Impl", marker='o')
plt.plot(df2["duration_sec"].values, label="Second Impl", marker='x')
plt.title("Goal Durations")
plt.xlabel("Goal Index")
plt.ylabel("Duration (sec)")
plt.legend()
plt.grid(True)
plt.show()
```

#  Performance Comparison Report of TheTwo Implementations


## 1. Hypotheses

To evaluate and compare the performance of two robotic goal-reaching implementations, we formulate the following hypotheses:

- **Null Hypothesis (H₀)**: There is no significant difference in the goal completion durations between Implementation 1 and Implementation 2.
- **Alternative Hypothesis (H₁)**: There is a significant difference in the goal completion durations, specifically that **Implementation 1 is faster** than Implementation 2.


## 2. Experimental Setup

### Objective
To compare the performance of two different implementations of a robot navigation system by measuring how long the robot takes to reach a series of fixed goals.

### Methodology
- A **fixed set of 40 goal coordinates** was used for both experiments to ensure consistency and fair comparison.
- The robot starts from the origin before attempting each goal.
- Each goal was sent and tracked individually using a ROS-based action client.
- The time taken (in seconds) from goal dispatch to successful arrival was recorded.

### Repetitions
- The experiment was conducted **twice**:
  - Once using **Implementation 1**
  - Again using **Implementation 2**
- Both trials used the **exact same goals** in the same order.


## 3. Results

| Metric                      | Implementation 1 | Implementation 2 |
|----------------------------|------------------|------------------|
| Number of Goals            | 40               | 40               |
| Successes                  | 40               | 40               |
| Failures                   | 0                | 0                |
| Average Duration (sec)     | 18.23            | 17.55            |

### t-test Results:
- **T-statistic**: 0.1605  
- **p-value**: 0.8729  

> Since the p-value is much greater than 0.05, the difference in durations is **not statistically significant**.


## 4. Discussion

The experimental data shows that:

- Both implementations successfully completed all goals.
- The mean duration is slightly lower for Implementation 2, but the difference is very small.
- Statistical analysis using a two-sample t-test confirms that **this difference is not significant**, with a p-value of 0.8729 (>> 0.05 threshold).

The outcome suggests that, from a performance standpoint, **both implementations are nearly equivalent in terms of time efficiency** under the tested conditions.


## 5. Conclusion

Based on the data and statistical analysis:

- The **null hypothesis is not rejected**.
- There is **no significant evidence** that Implementation 1 performs better (faster) than Implementation 2.
- Thus, we conclude that **both implementations perform similarly** in terms of goal-reaching duration.
